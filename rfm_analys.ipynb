# Import Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Mengambil data melalui URL
url = 'https://docs.google.com/spreadsheets/d/1rbLQO-YDTrfhA9pw7NuNs-e2n-nPKdkr/edit?usp=sharing&ouid=103893180789515944577&rtpof=true&sd=true'
url = 'https://drive.google.com/uc?id=' + url.split('/')[-2]

df = pd.read_excel(url)
df
```

```python
df.info()
```

# Data Preprocessing

## Mengambil Kolom yang Digunakan

```python
data = df[['Nama', 'Nama Event', 'Waktu', 'Benefit Total Mobo']]
data
```

```python
# Buat daftar nama unik di kolom 'Nama'
unique_names = data['Nama'].unique()

# Buat mapping dari nama unik ke abjad (A, B, C, ...)
name_mapping = {name: chr(65 + i) for i, name in enumerate(unique_names)}

# Ganti nilai kolom 'Nama' dengan abjad menggunakan mapping
data['Nama'] = data['Nama'].map(name_mapping)

# Tampilkan hasil
data
```

## Cek Missing Value

```python
data.info()
```

## Mengubah Kolom Waktu ke Dalam Formate Datetime

```python
# Parsing kolom 'Waktu' dengan format eksplisit
data['Waktu'] = pd.to_datetime(data['Waktu'], format='%d/%m/%Y %H.%M.%S', errors='coerce')
data.info()
```

```python
data.sort_values(by='Waktu')
```

# RFM Analyst

## Menghitung Recency

```python
df_recency = data.groupby(by='Nama', 
						as_index=False)['Waktu'].max()
df_recency.columns = ['Nama', 'LastEventDate']
recent_date = df_recency['LastEventDate'].max()
df_recency['Recency'] = df_recency['LastEventDate'].apply(
	lambda x: (recent_date - x).days)
df_recency.sort_values(by='Recency')
```

## Menghitung Frekuensi

```python
frequency_df = data.drop_duplicates().groupby(
	by=['Nama'], as_index=False)['Waktu'].count()
frequency_df.columns = ['Nama', 'Frequency']
frequency_df.sort_values(by='Frequency')
```

## Menghitung Monetary

```python
monetary_df = data.groupby(by='Nama', as_index=False)['Benefit Total Mobo'].sum()
monetary_df.columns = ['Nama', 'Monetary']
monetary_df.sort_values(by='Monetary')
```

```python
rf_df = df_recency.merge(frequency_df, on='Nama')
rfm_df = rf_df.merge(monetary_df, on='Nama').drop(
	columns='LastEventDate')
rfm_df
```

```python
rfm_df['R_rank'] = rfm_df['Recency'].rank(ascending=False)
rfm_df['F_rank'] = rfm_df['Frequency'].rank(ascending=True)
rfm_df['M_rank'] = rfm_df['Monetary'].rank(ascending=True)

# normalizing the rank of the customers
rfm_df['R_rank_norm'] = (rfm_df['R_rank']/rfm_df['R_rank'].max())*100
rfm_df['F_rank_norm'] = (rfm_df['F_rank']/rfm_df['F_rank'].max())*100
rfm_df['M_rank_norm'] = (rfm_df['M_rank']/rfm_df['M_rank'].max())*100

rfm_df.drop(columns=['R_rank', 'F_rank', 'M_rank'], inplace=True)

rfm_df
```

## Menghitung RFM Score

```python
rfm_df['RFM_Score'] = 0.15*rfm_df['R_rank_norm']+0.28 * \
	rfm_df['F_rank_norm']+0.57*rfm_df['M_rank_norm']
rfm_df['RFM_Score'] *= 0.05
rfm_df = rfm_df.round(2)
rfm_df[['Nama', 'RFM_Score']].sort_values(by='RFM_Score')
```

## Klasterisasi Agent

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np

# Menyiapkan data untuk clustering (menggunakan kolom 'RFM_Score')
data = rfm_df[['RFM_Score']]

# Menentukan rentang jumlah klaster
wcss = []  # Within-Cluster Sum of Squares
k_values = range(1, 11)

# Menghitung WCSS untuk setiap jumlah klaster
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)

# Membuat visualisasi elbow
plt.figure(figsize=(8, 5))
plt.plot(k_values, wcss, marker='o', linestyle='--', color='b')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS')
plt.xticks(k_values)
plt.grid()
plt.show()
```

```python
# Mengelompokkan agen menjadi 3 bagian berdasarkan RFM_Score
rfm_df["Agent_segment"] = np.where(
    rfm_df['RFM_Score'] > 4, "Top Agent",
    np.where(rfm_df['RFM_Score'] > 2.5, "Medium Agent", "Low Agent")
)

# Menampilkan kolom Nama, RFM_Score, dan Agent_segment
rfm_df[['Nama', 'RFM_Score', 'Agent_segment']].sort_values(by='RFM_Score')
```
